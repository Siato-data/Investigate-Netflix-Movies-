```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

```

```python
# Start coding here! Use as many cells as you like
print(netflix_df.info())
```

We keep only movies release between 1990 and 1999

```python
# affichont les differentes data de la variable type
x=[]
for i in netflix_df['type']:
    if i not in x:
        x.append(i)

print(x)

#print(netflix_df['type'].unique())
```

```python
#Now we put in a variable movie only Movie's type 
movie = netflix_df[netflix_df['type'] == 'Movie' ]
print(movie)
```

```python
#and we keep only the movie made in 90' Era
movie_after90 = movie[movie['release_year']>=1990 ]
movie_90 = movie_after90[movie_after90['release_year']<=1999]
print(movie_90)
```

Time to find the most frequent movie duration now

```python
#Most frequent movie duration in 1950
print(movie_90['duration'].mode())
duration = 94
```

```python
#faire un graphique pour la visualisation 
bins=[25,37.5,50,62.5,75,87.5,100,112.5,125,137.5,150,162.5,175,187.5,200]
plt.hist(movie_90['duration'],bins=bins, edgecolor = 'black')
plt.xticks(rotation=45)
plt.title('Visualisation of duration of movie', fontweight='bold')
plt.xlabel('movie duration ',color='darkblue')
plt.ylabel('number of movie',color='darkblue')
plt.tight_layout()
plt.show()
```

Now we will found the short action movies !

```python
#Affichons les diffentes data de la variable genre 
genres=[]
for i in movie_90['genre']:
    if i not in genres:
        genres.append(i)
print(genres)
```

```python
#Number of short action movies count
action_90 = movie_90[movie_90['genre']=='Action']
short_movie = action_90[action_90['duration']<= 90 ]
short_movie_count = short_movie.count()
print(short_movie_count)
```
